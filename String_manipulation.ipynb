Pick a FASTA sequence in the NCBI and import it to Colab. Read the file removing the header and use functions DNA module to calculate the GC content.

# Exercise 2

Pick a PDB protein file, read it and calculate the number of atoms of glycine, aspartate and leucine.

```python
glycine = 0
aspartate = 0
leucine = 0

with open('7ef8.pdb', 'r') as protein:
  for line in protein:
    if 'GLY' in line:
      glycine += 1
    if 'ASP' in line:
      aspartate += 1
    if 'LEU' in line:
      leucine += 1

print(f'There are {glycine} atoms of glycine.')
print(f'There are {aspartate} atoms of aspartate.')
print(f'There are {leucine} atoms of leucine.')
```

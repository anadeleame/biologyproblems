# Exercise 1

Write a regex that extracts abbreviated scientific names of a string. Ex: *H. sapiens*.

```python
import re
phrase = '''
Of the species closest to H. sapiens currently alive, we can mention P. troglodytes and P. paniscus,
which still present notable differences at the most diverse levels.
'''
re.findall('[A-Z]\.\s[a-z]+', phrase)
```

# Exercise 3

Huntington's disease is a genetic disease which is part of the group of the polyglutamine diseases (polyQ). A normal sequence, which does not cause the disease, generally has between 6 and 35 repetitions of the sequence of codons that encode glutamine (CAA/CAG). A number greater than 35 repetitions of these codons practically guarantees the establishment of the disease. Write a program that, with the help of regular expressions, identifies whether in an RNA sequence entered by the user, there is a sequence with 35 times or more of the codons that code for glutamine. **OBS**: the sequence can have 35 CAG's in repetition, 35 CAA's in repetition, or it can have several CAA's and CAG's mixed in the sequence appearing 35 times or more.

```python
sequence = str(input('Type your RNA sequence: '))

hunt = re.search('(CAG|CAG){35,}', sequence)

if hunt == None:
  print(f'Your sequence does not have a pattern of Huntingon.')
else:
  print(f'Your sequence has a pattern of Huntington.')

```

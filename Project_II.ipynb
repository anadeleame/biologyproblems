### Create a module that calculates the gene frequencies of a population and returns whether the frequencies are expected or not. This is made by the chi-square test.
**Level of significance: 0.95**

**Degree of freedom: 1**

▶ Step 1 = calculate alleles:

**P =** (2*PP observed + PQ observed) / 2 * total

**Q =** (2*QQ observed + PQ observed) / 2 * total


▶ Step 2 = calculate expected genotypes through Hardy-Weinberg's formula:

**Pexp =** P² * total

**PQexp =** 2*P*Q * total

**Qexp =**  Q² total

▶ Step 3 = calculate chi-square test comparing observed x expected genotypes. If the result is bigger than 3,841, there is a genotype imbalance in this population; if not, it is balanced.

```python
#The module created on Spyder:

def chisquare():
    '''
    Calculates chi-square to compare observed and expected genotypes.
    '''
    indPP = int(input('Type the number of PP individuals: '))
    indPQ = int(input('Type the number of PQ individuals: '))
    indQQ = int(input('Type the number of QQ individuals: '))
    total_ind = indPP + indPQ + indQQ

    Pobs = (2 * indPP + indPQ) / (2 * total_ind)
    Qobs = (2 * indQQ + indPQ) / (2 * total_ind)

    observed = [indPP, indPQ, indQQ]

    expecP = (Pobs ** 2) * total_ind
    expecPQ = (2 * Pobs * Qobs) * total_ind
    expecQ = (Qobs ** 2) * total_ind

    expected = [expecP, expecPQ, expecQ]

    result = 0

    for obs, exp in zip(observed,expected):
        result += (obs-exp)**2/exp

    print(result)
```
